### Converting RGB image to Grayscale image

```python
from skimage import color
from skimage import io
```

```python
image = color.rgb2gray(io.imread('new_image.png'))
plt.imshow(image,cmap=plt.cm.gray)
```

```python
image.shape
```

```python
rows=image.shape[0]
cols=image.shape[1]
```

### Reshaping the image

```python
image_new=image.reshape(rows*cols,1)
```

```python
from sklearn.cluster import KMeans
```

```python
kmean=KMeans(n_clusters=5)
kmean.fit(image_new)
```

```python
labels=kmean.labels_
```

```python
label=labels.reshape(rows,cols)
```

```python
plt.imshow(image,cmap=plt.cm.gray)
```

```python
plt.imshow(label,cmap=plt.cm.gray)
```

### The quality of the compressed image with 5 clusters is very poor when compared to the original image
