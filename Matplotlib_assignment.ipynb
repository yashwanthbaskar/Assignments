Transform iris data into 3 dimensions  and plot a 3d chart with transformed dimensions and colour each data  point with specific class. 


```python
import numpy as np 
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D 
from sklearn.decomposition import PCA
from sklearn import datasets 
```

```python
df=datasets.load_iris()
```

```python
df.data
```

```python
df['species']
```

```python
y=df['target']
y
```

```python
pca = PCA(n_components=3)
x = pca.fit_transform(df.data)
```

```python
fig=plt.figure(111,figsize=(12,8))
axes=Axes3D(fig,elev=-150,azim=200)
```

```python
x_col1=x[:,0]
```

```python
x_col2=x[:,1]
```

```python
x_col3=x[:,2]
```

```python
axes.scatter(x_col1,x_col2,x_col2,c=y,cmap='Blues')
```

```python
plt.show()
```

```python
axes.set_title('IRIS DATA')
axes.set_xlabel('X AXIS')
axes.set_ylabel('Y AXIS')
axes.set_zlabel('Z AXIS')
```
